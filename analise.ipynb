```python
#PAGAMENTO
import plotly.express as px
import pandas as pd
import math as mt
tabela= pd.read_excel("BaseCancelamento.xlsx")
tabela=tabela.drop(columns=["Código_Cliente"])
analise_cancelamento_x_pagamento=tabela.groupby(["Forma de Pagamento","Cancelou"]).size().reset_index(name="Quantidade")
grafico=px.bar(analise_cancelamento_x_pagamento, x="Forma de Pagamento", y="Quantidade", color="Cancelou", barmode="group", title="Análise de Cancelamento por Forma de Pagamento")
grafico.show()
soma=tabela["Cancelou"].value_counts()
display(soma)
# Adicionando a coluna de porcentagem com base no total para cada valor de Cancelou
analise_cancelamento_x_pagamento["Porcentagem"] = analise_cancelamento_x_pagamento.apply(lambda linha: linha["Quantidade"] / soma[linha["Cancelou"]],axis=1)
lista_de_valores=[]
for porcentagem in analise_cancelamento_x_pagamento["Porcentagem"]:
    lista_de_valores.append(porcentagem)
subtracao_boleto=mt.fabs(lista_de_valores[0]-lista_de_valores[1])
subtracao_cartao=mt.fabs(lista_de_valores[2]-lista_de_valores[3])
subtracao_pix=mt.fabs(lista_de_valores[4]-lista_de_valores[5])
value_of_info_pagamento=(subtracao_boleto + subtracao_cartao + subtracao_pix)
value=pd.DataFrame({"Valor da informação":[value_of_info_pagamento]})
display(value)
    





```

```python
#SUPORTE
import plotly.express as px
import pandas as pd
import math as mt
tabela= pd.read_excel("BaseCancelamento.xlsx")
tabela=tabela.drop(columns=["Código_Cliente"])
analise_cancelamento_x_interacaosuporte=tabela.groupby(["Interacoes_Suporte","Cancelou"]).size().reset_index(name="relação")
soma=tabela["Cancelou"].value_counts()
display(soma)
analise_cancelamento_x_interacaosuporte["porcentagem"]= analise_cancelamento_x_interacaosuporte.apply(lambda linha: linha["relação"] / soma[linha["Cancelou"]],axis=1)
analise_cancelamento_x_avaliacaosuporte=tabela.groupby(["Avaliação_Suporte","Cancelou"]).size().reset_index(name="avaliação")
analise_cancelamento_x_avaliacaosuporte["avaliação\porcentagem"]= analise_cancelamento_x_avaliacaosuporte.apply(lambda  linha: linha["avaliação"] / soma[linha["Cancelou"]],axis=1)
grafico2=px.bar(analise_cancelamento_x_avaliacaosuporte, x="Avaliação_Suporte", y="avaliação", color="Cancelou", barmode="group", title="Análise de Cancelamento por Avaliação de Suporte")
grafico2.show()
grafico=px.bar(analise_cancelamento_x_interacaosuporte, x="Interacoes_Suporte", y="relação", color="Cancelou", barmode="group", title="Análise de Cancelamento por Interações com Suporte")
grafico.show()
lista_de_valores=[]
for porcentagem in analise_cancelamento_x_interacaosuporte["porcentagem"]:
    lista_de_valores.append(porcentagem)
avaliacao=[]
for i in analise_cancelamento_x_avaliacaosuporte["avaliação\porcentagem"]:
    avaliacao.append(i)
lista_de_avaliacao=[mt.fabs(avaliacao[i] - avaliacao[i + 1]) for i in range(0, len(avaliacao), 2)]
lista_de_cancelou_renovou = [mt.fabs(lista_de_valores[i] - lista_de_valores[i + 1]) for i in range(0, len(lista_de_valores), 2)]
valor_of_info_interacao_x_avaliacao=pd.DataFrame({"valor da informação":[sum(lista_de_cancelou_renovou)], "valor da avaliação":[sum(lista_de_avaliacao)]})
display(valor_of_info_interacao_x_avaliacao)












```

```python
#ONBORDING
import plotly.express as px
import pandas as pd
import math as mt
tabela= pd.read_excel("BaseCancelamento.xlsx")
tabela=tabela.drop(columns=["Código_Cliente"])
analise_cancelamento_x_onbording=tabela.groupby(["Experiencia_OnBoarding", "Cancelou"]).size().reset_index(name="quantidade")
soma=tabela["Cancelou"].value_counts()
display(soma)
analise_cancelamento_x_onbording["porcentagem"]=analise_cancelamento_x_onbording.apply(lambda linha: linha["quantidade"]/soma[linha["Cancelou"]], axis=1)
grafico=px.bar(analise_cancelamento_x_onbording, x="Experiencia_OnBoarding",y="quantidade",color="Cancelou",title="analise de cancelamento por experiência de onboarding",barmode="group")
grafico.show()
lista_de_valores=[]
for porcentagem in (analise_cancelamento_x_onbording["porcentagem"]):
    lista_de_valores.append(porcentagem)
lista_de_subtracao = [mt.fabs(lista_de_valores[i] - lista_de_valores[i + 1]) for i in range(0, len(lista_de_valores), 2)]
tab=pd.DataFrame({"Experiencia_OnBoarding":tabela["Experiencia_OnBoarding"].unique(), "valor": lista_de_subtracao})
grafico2=px.bar(tab,x="Experiencia_OnBoarding",y="valor",title="valor de influência das notas", barmode="group")
grafico2.show()
valor_of_info_onbording=pd.DataFrame({"valor da informação":[sum(lista_de_subtracao)]})
display(valor_of_info_onbording)

```

```python
#AVALIAÇÃO DA PLATAFORMA
import plotly.express as px
import pandas as pd
import math as mt
tabela= pd.read_excel("BaseCancelamento.xlsx")
tabela=tabela.drop(columns=["Código_Cliente"])
analise_paltaforma=tabela.groupby(["Avaliação_da_Plataforma","Cancelou"]).size().reset_index(name="Quantidade")
soma=tabela["Cancelou"].value_counts()
display(soma)
analise_paltaforma["Porcentagem"]=analise_paltaforma.apply(lambda linha: linha["Quantidade"]/soma[linha["Cancelou"]], axis=1)
grafico=px.bar(analise_paltaforma, x="Avaliação_da_Plataforma", y="Quantidade", color="Cancelou", barmode="group", title="Análise de Cancelamento por Avaliação da Plataforma")
grafico.show()
lista_de_valores=[]
for porcentagem in analise_paltaforma["Porcentagem"]:
    lista_de_valores.append(porcentagem)
lista_de_subtracao = [mt.fabs(lista_de_valores[i] - lista_de_valores[i + 1]) for i in range(0, len(lista_de_valores), 2)]
valor_of_info_plataforma=pd.DataFrame({"Valor da Informação":[sum(lista_de_subtracao)]})
display(valor_of_info_plataforma)
```
